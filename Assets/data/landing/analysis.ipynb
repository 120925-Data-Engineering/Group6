```python
import pandas as pd     # data analysis
import json     # json parsing library
import os   # handling file paths and folders
```

```python
# Creates list to store raw records
data = [] 


with open('transaction_events_1767986916.1293132.json', 'r') as f:
    for line in f:
        # Parse each line and add to data list
        data.append(json.loads(line))


# takes out individual products and pair them with their transaction info???
df = pd.json_normalize(data, 'products', ['transaction_id', 'transaction_type'])


# Sums quantities specifically for "purchase" events
total_purchased = df[df['transaction_type'] == 'purchase']['quantity'].sum()


# shows result nicely (simple)
print("-" * 50)
print("               TRANSACTION SUMMARY   ")
print("-" * 50)
print(f"Status:       Success")
print(f"File Processed: transaction_events_...132.json")
print(f"Total Items:    {total_purchased}")
print("-" * 50)
```

```python

# Filter the data to only include completed purchases
purchase_df = df[df['transaction_type'] == 'purchase'].copy()


# prints our summary neatly
print("-" * 100)
print("                                    TRANSACTION ANALYSIS REPORT")
print("-" * 100)
print(f"Total Unique Transactions: {purchase_df['transaction_id'].nunique()}")
print(f"Total Quantity Sold:       {purchase_df['quantity'].sum()}")
print("-" * 100)


# Select the columns we want displayed
report_columns = ['product_name', 'category', 'quantity', 'unit_price']
final_report = purchase_df[report_columns]


# Print the detailed list of items purchased - shows the first 16 results and the last 4 (detailed)
print("\nItemized Purchase Details:")
print(final_report.to_string(index=False))
```
